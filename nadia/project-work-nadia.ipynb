```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
per_capita_alcohol = "Resources/per-capita-alcohol-1890.csv"
mental_health_data = "Resources/share-with-mental-and-substance-disorders.csv"
happiness_data = "Resources/happiness-cantril-ladder.csv"
wine_consumption = "Resources/wine-consumption-per-person.csv"
alcohol_expenditure = "Resources/alcohol-expenditure-as-share-of-total.csv"

per_capita_alcohol_df = pd.read_csv(per_capita_alcohol)
mental_health_data_df = pd.read_csv(mental_health_data)
happiness_data_df = pd.read_csv(happiness_data)
wine_consumption_df = pd.read_csv(wine_consumption)
alcohol_expenditure_df = pd.read_csv(alcohol_expenditure)
```

```python
alcohol_and_mental_health_df = pd.merge(per_capita_alcohol_df, mental_health_data_df, how='outer', on=['Entity', 'Code', 'Year'])
alcohol_and_mental_health_df.sample(10)
```

```python
alcohol_mental_health_happiness_df = pd.merge(alcohol_and_mental_health_df, happiness_data_df, how='outer', on=['Entity', 'Code', 'Year'])
alcohol_mental_health_happiness_df.sample(10)
```

```python
alcohol_mental_health_happiness_wine_df = pd.merge(alcohol_mental_health_happiness_df, wine_consumption_df, how='outer', on=['Entity', 'Code', 'Year'])
alcohol_mental_health_happiness_wine_df.sample(10)
```

```python
alcohol_expense_mental_health_happiness_wine_df = pd.merge(alcohol_mental_health_happiness_wine_df, alcohol_expenditure_df, how='outer', on=['Entity', 'Code', 'Year'])
alcohol_expense_mental_health_happiness_wine_df
```

```python
alcohol_expense_mental_health_happiness_wine_df=alcohol_expense_mental_health_happiness_wine_df.dropna(axis='rows')
alcohol_expense_mental_health_happiness_wine_df
```

```python
renamed_df=alcohol_expense_mental_health_happiness_wine_df.rename(columns=({"Alcohol consumption since 1890 (Alexander & Holmes, 2017)": "Alcohol Consumption (Liters)", "Prevalence - Mental and substance use disorders - Sex: Both - Age: Age-standardized (Percent)" : "Mental Health Disorders(%)", "Life satisfaction in Cantril Ladder (World Happiness Report 2019)" : "Happiness Score", "Indicator:Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol) - Beverage Types:Wine": "Wine Consumption Per Capita (Liters)", "Indicator:Alcohol expenditure as a per cent of total household expenditure" : "Alcohol Expenditure (%)" })).reset_index()
renamed_df
```

```python
wine_and_mental_health_df = pd.merge(wine_consumption_df, mental_health_data_df, how='outer', on=['Entity', 'Code', 'Year'])
wine_and_mental_health_df=wine_and_mental_health_df.dropna(axis='rows')
wine_and_mental_health_df=wine_and_mental_health_df.rename(columns={"Indicator:Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol) - Beverage Types:Wine":"Wine Consumption (in litres)", "Prevalence - Mental and substance use disorders - Sex: Both - Age: Age-standardized (Percent)":"Prevalence of Mental Disorders (%)"})
wine_and_mental_health_df.sample(35)
```

```python
grouped_wine_and_mental_health=wine_and_mental_health_df.groupby("Entity")[["Prevalence of Mental Disorders (%)"]].mean()
grouped_wine_and_mental_health
```

```python
grouped_wine_and_mental_health.plot(kind='barh', figsize=(12,3), title ='Mental Disorders per Country')
```

```python
wine_and_mental_health_df.boxplot("Prevalence of Mental Disorders (%)", by="Wine Consumption (in litres)", figsize=(20, 10))
```

```python
plt.figure(figsize=(15,3))
plt.barh(grouped_wine_and_mental_health)
plt.xlabel('Country')
plt.ylabel('Mental Health Prevalence (%)')
plt.title('Mental Health Prevalence by Country')
plt.show()
```
